```python
import pandas as pd
from konlpy.tag import Hannanum
from konlpy.tag import Okt
import pyspark as sc
from pyspark.sql import SparkSession
import csv

file_path = r'clustering_6_corpus.csv'
df = pd.read_csv(file_path, encoding='utf-8-sig', header=None)
df = df.iloc[1:]

for i , row in df.iterrows():
    n = int(row[3])
    for k in range(0, 6):
        if n == k:
            file_path = r'clustering_6_corpus_' + str(n) + '.csv'
            pd.DataFrame([{
                'contents' : row[2]
                }]).to_csv(file_path, encoding='utf-8-sig', mode='a', index=False, header=False)
```

```python
import time
from konlpy.tag import Okt
from tqdm import tqdm

stopwords = []
with open(r'stopword.txt', 'r', encoding='utf-8-sig') as f:
    txt = csv.reader(f)
    for word in txt:
        stopwords.append(word)

okt = Okt()

for i in range(0,6):
 file_path = r'clustering_6_corpus_' + str(i) + '.csv'
 df = pd.read_csv(file_path, encoding='utf-8-sig', header=None) 

 noun_frequency = {}
 total_time = 0
 iteration_count = 0
 for k, row in df.iterrows():
    content = str(row)
    start_time = time.time()
    nouns = []

    try:
        nouns = okt.nouns(content)
    except UnicodeDecodeError:
        continue
    except Exception as e:
        print(e)
        continue

    for noun in nouns:
        if len(noun) > 1 and noun not in stopwords:
            if noun in noun_frequency:
                noun_frequency[noun] += 1
            else:
                noun_frequency[noun] = 1

    end_time = time.time()
    total_time += end_time - start_time
    iteration_count += 1

 average_time = total_time / iteration_count
 average_speed = 1 / average_time

 print(f'\nAverage iteration speed: {average_speed:.2f}it/s')
 print(str(i) + 'th label words frequency')
 top_nouns = sorted(noun_frequency.items(), key=lambda x: x[1], reverse=True)[:100]

 for noun in top_nouns:
     print(noun)
```
